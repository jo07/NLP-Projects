```python
hamlet = nltk.corpus.gutenberg.words( 'shakespeare-hamlet.txt')
hamlet
```

```python
" ".join(hamlet[:100])
```

```python
hamlet_500 = " ".join(hamlet[:500])
```

```python
from nltk.tokenize import word_tokenize
```

```python
token_hamlet_500 = word_tokenize(hamlet_500)
token_hamlet_500
```

```python
# get word count of the tokes
from nltk.probability import FreqDist
fdist = FreqDist()
```

```python
for word in token_hamlet_500:
    fdist[word.lower()] += 1
fdist
```

```python
fdist_top10 = fdist.most_common(10)
fdist_top10
```

```python
# blank tokenizer
from nltk.tokenize import blankline_tokenize
blank_hamlet_500 = blankline_tokenize(hamlet_500)
blank_hamlet_500
```

```python
len(blank_hamlet_500)
```

```python
moby_dick = nltk.corpus.gutenberg.words( 'melville-moby_dick.txt')
moby_dick
```

```python
moby_dick_500 = " ".join(moby_dick[:500])
moby_dick_500
```

```python
blank_moby_dick_500 = blankline_tokenize(moby_dick_500)
blank_moby_dick_500
```

```python
len(blank_moby_dick_500)
```

```python
from nltk.util import bigrams, trigrams, ngrams
```

```python
token_moby_dick_500 = word_tokenize(moby_dick_500)
token_moby_dick_500
```

```python
list(bigrams(token_moby_dick_500))
```

```python
list(trigrams(token_moby_dick_500))
```

```python
ngram_10 = list(ngrams(token_moby_dick_500, n = 10))
```

```python
[ " ".join(word) for word in ngram_10]
```
